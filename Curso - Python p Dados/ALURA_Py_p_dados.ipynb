#### Exemplo 1

*   Separar uma lista de str e flt em duas.

```python
notas_turma_ex5 = ['Jo√£o', 8.0, 9.0, 10.0, 'Maria', 9.0, 7.0, 6.0, 'Jos√©', 3.4, 7.0, 7.0, 'Cl√°udia', 5.5, 6.6, 8.0, 'Ana', 6.0, 10.0, 9.5]
```

```python
nomes = []
notas_juntas = []

for i in range(len(notas_turma_ex5)):
  if i % 4 ==0:
      nomes.append(notas_turma_ex5[i])
  else:
      notas_juntas.append(notas_turma_ex5[i])
```

```python
notas = []

for i in range(0, len(notas_juntas), 3):
  notas.append([notas_juntas[i], notas_juntas[i+1], notas_juntas[i+2]])
notas
```

#### Exemplo 2

*   Criar uma lista com nomes ao lado de seus respectivos c√≥digos.

```python
estudantes = ["Jo√£o", "Maria", "Jos√©", "Cl√°udia", "Ana"]
```

```python
def gera_codigo():
  return str(randint(0,999))
```

```python
codigo_estudantes = []

for i in range(len(estudantes)):
  codigo_estudantes.append((estudantes[i], estudantes[i][0] + gera_codigo()))

codigo_estudantes
```

#### Exemplo 3

*   Extrair informa√ß√µes de uma lista de tuplas.

```python
nomes = [('Jo√£o', 'J720'), ('Maria', 'M205'), ('Jos√©', 'J371'), ('Cl√°udia', 'C546'), ('Ana', 'A347')]
medias = [9.0, 7.3, 5.8, 6.7, 8.5]
```

```python
lista_nomes = [nome[0] for nome in nomes]
lista_nomes
```

```python
estudantes = list(zip(lista_nomes, medias))
estudantes
```

```python
candidatos = [estudante[0] for estudante in estudantes if estudante[1] >= 8]
candidatos
```
