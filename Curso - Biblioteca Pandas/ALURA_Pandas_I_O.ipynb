## 05 - Trabalhando com banco de dados ðŸŽ²





> Na primeira aula importamos a biblioteca sqlalchemy, junto a algumas funÃ§Ãµes desta biblioteca, e criamos o "motor" do banco de dados manualmente. Na segunda aula vimos como escrever dentro do nosso banco de dados, importando dados csv e utilizando to_sql para salvar uma dada tabela em nosso banco. Na terceira aula, aprendemos como fazer uma consulta(query) em SQL(SELECT FROM WHERE) e como ler esta com read_sql, salvar essa tabela da consulta e ler esta tabela com read_sql_table. Na quarta aula, aprendemos como alterar dados que estavam no banco, usando DELETE, UPDATE, SET, juntamente com mÃ©todo with.

#### Exemplo 1


*   Escrevendo no banco de dados âœï¸

```python
dados = pd.read_csv(url_4)
dados.head()
```

```python
dados.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine)
print(inspector.get_table_names())
```

#### Exemplo 2


*   Consultando e lendo sql ðŸ”

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda="Empregado"'
```

```python
empregados = pd.read_sql(query, engine)
empregados
```

```python
#salvando
empregados.to_sql('empregados', con=engine, index=False)
```

```python
#o read_sql permite ler uma consulta e o read_sql_table uma tabela completa.
pd.read_sql_table('empregados', engine)
```

#### Exemplo 3


*   Alterando informaÃ§Ãµes ðŸ“

```python
query = 'SELECT * FROM clientes'
```

```python
pd.read_sql(query, engine)
```

```python
query = 'DELETE FROM clientes WHERE ID_Cliente=5008804'
with engine.connect() as conn:
    conn.execute(sqlalchemy.text(query))
    conn.commit()
```

```python
pd.read_sql_table('clientes', engine)
```

```python
query = 'UPDATE clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808'
with engine.connect() as conn:
    conn.execute(sqlalchemy.text(query))
    conn.commit()
```

```python
pd.read_sql_table('clientes', engine)
```
